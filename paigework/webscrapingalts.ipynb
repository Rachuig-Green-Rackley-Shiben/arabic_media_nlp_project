```python
import requests
from bs4 import BeautifulSoup as bs
import re

import pandas as pd
```

```python
URL = 'https://english.alarabiya.net/tools/search?query=america'
  
req = requests.get(URL)
soup = bs(req.text, 'html.parser')
  
titles = soup.find_all('a', class_= 'latest_link')
  
print(titles[4].text)
```

```python
# gives us the urls to the 500 page search.

page = 1
titles = []
while page != 500:
      url = f"https://english.alarabiya.net/tools/search?query=america$page={page}"
      print(url)
      page = page + 1
```

```python
#trying to build off of that

page = 1
titles = []
while page != 500:
      url = f"https://english.alarabiya.net/tools/search?query=america$page={page}"
      response = requests.get(url)
      html = response.content
      soup = bs(html, "lxml")
      for h3 in soup.find_all("a", class_="latest_link"):
            titles.append(h3.get_text(strip=True))
      page = page + 1
```
