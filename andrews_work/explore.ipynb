# Hypothesis testing

$H_{0}$: Calculated article sentiment is independant of calculated headline sentiment

$H_{a}$: Calculated article sentiment is NOT independant of calculated headline sentiment

```python
alpha = .05
observed = pd.crosstab(encoded.text_label, encoded.headline_label)
observed
```

```python
from scipy import stats
chi2, p, degf, expected = stats.chi2_contingency(observed)
```

```python
chi2, p
```

```python
pd.DataFrame(expected)
```

```python
p < alpha
```

Because p is less than alpha we can reject the Null Hypothesis and determine that calculated article sentiment is NOT independant of calculated headline sentiment.
