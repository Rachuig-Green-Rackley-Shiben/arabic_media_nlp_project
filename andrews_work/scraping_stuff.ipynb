```python
from requests import get
from bs4 import BeautifulSoup
import re
```

```python
url = 'https://english.alarabiya.net/tools/search?query=america&pageNo=1'

response = get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup.select('a.latest_link')
```

```python
link_list = []
for i in range(len(soup.select('a.latest_link'))):
    text = str(soup.select('a.latest_link')[i].get_text)
    link = re.search(r'href="(.+)"', text)
    link_list.append('https://english.alarabiya.net/'+link[1])
```
